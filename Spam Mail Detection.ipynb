```python
import pandas as pd 
import numpy as np
import tensorflow as tf 
```

```python
import csv
```

```python
with open('spam2.csv','r') as csv_file:
    csv_reader=csv.DictReader(csv_file) #store the contents in a dictionary
    data=[] #initialize a lsit to store all dictionary items
    for row in csv_reader:
        data.append(row)  
```

```python
import json
with open('spam.json','w') as json_file:
    json.dump(data, json_file, indent=2)  
```

```python
with open('spam.json') as file:
    data_new=json.load(file)
sentences=[]
label=[]
for item in data_new:
    sentences.append(item['v2'])
    label.append(item['v1'])
```

```python
label=np.array(label)
labels=[]
for x in range(len(label)):
    if label[x]=='ham':
        labels.append(0)
    else:
        labels.append(1) 
```

```python
sentences=np.array(sentences)
labels=np.array(labels)
```

```python
vocab_size=10000
embedded_dim=16
max_length=35
padding_type='post'
oov_token='<OOV>'
train_size=int(len(sentences)*0.8)

train_sentences=sentences[:train_size]
train_labels=labels[:train_size]
test_sentences=sentences[train_size:]
test_labels=labels[train_size:]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
tokenizer=Tokenizer()
tokenizer.fit_on_texts(train_sentences)
word_index=tokenizer.word_index
print(word_index)
```

```python
sequences=tokenizer.texts_to_sequences(train_sentences)
padded=pad_sequences(sequences, padding=padding_type, maxlen=max_length,truncating='pre')
```

```python
print(sequences)
print(padded)
```

```python
test_sequences=tokenizer.texts_to_sequences(test_sentences)
test_padded=pad_sequences(test_sequences, padding=padding_type, maxlen=max_length,truncating='pre')
```

```python
model=tf.keras.models.Sequential([tf.keras.layers.Embedding(vocab_size, embedded_dim, input_length=max_length),
                                 tf.keras.layers.Flatten(),
                                 tf.keras.layers.Dense(64,activation='relu'),
                                 tf.keras.layers.Dense(1,activation='sigmoid')])
model.build()
```

```python
model.summary()
```

```python
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
             optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
             metrics=['accuracy'])
history=model.fit(padded, train_labels,epochs=20,validation_data=(test_padded,test_labels))
```

```python
import matplotlib.pyplot as plt
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']
epochs=range(len(acc))

plt.plot(epochs, acc)
plt.plot(epochs, val_acc)
plt.title('Training vs validation accuracy')
plt.figure()

plt.plot(epochs, loss)
plt.plot(epochs, val_loss)
plt.title('Training vs validation loss')
plt.figure()
```
